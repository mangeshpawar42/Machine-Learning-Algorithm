```python
import pandas as pd 
import numpy as np
from pandas_profiling import ProfileReport
import warnings
warnings.simplefilter("ignore")
```

```python
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
model = smf.ols(formula='Sales ~ TV',data=df).fit()
model.summary()
```

```python
model = smf.ols(formula='Sales ~ TV + Radio',data=df).fit()
model.summary()
```

```python
model = smf.ols(formula='Sales ~ Newspaper',data=df).fit()
model.summary()
```

```python
model = smf.ols(formula='Sales ~ Newspaper + Radio ',data=df).fit()
model.summary()
```

```python
model = smf.ols(formula='Sales ~ Newspaper + Radio + TV',data=df).fit()
model.summary()
```
